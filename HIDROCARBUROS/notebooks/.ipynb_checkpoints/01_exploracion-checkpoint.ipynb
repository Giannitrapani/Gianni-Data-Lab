```python
# ============================================================
# CELDA 1 - VERIFICAR INSTALACIONES
# ============================================================
# Con Anaconda ya todo esto viene instalado.
# Esta celda solo confirma que las versiones están OK.
# Si algo falla acá, ejecutar en terminal: pip install [nombre]
# ============================================================

import sys
print("🐍 Python version:", sys.version)

import requests
print("✅ requests:", requests.__version__)

import pandas
print("✅ pandas:", pandas.__version__)

import numpy
print("✅ numpy:", numpy.__version__)

print("")
print("🎉 Todo listo para trabajar!")
```

```python
# ============================================================
# CELDA 2 - IMPORTAR LIBRERÍAS
# ============================================================
# El 'as pd' y 'as np' son alias — convención universal en Data Science.
# En lugar de escribir pandas.DataFrame, escribís pd.DataFrame.
# Todos los Data Scientists del mundo usan estos mismos alias.
# ============================================================

import requests        # para hacer llamadas HTTP a la API
import pandas as pd    # para trabajar con tablas de datos (DataFrames)
import numpy as np     # para cálculos numéricos

print("📦 Librerías importadas correctamente")
```

```python
# ============================================================
# CELDA 3 - CONECTARSE A LA API DEL MINISTERIO DE ENERGÍA
# ============================================================
# La API es pública, gratuita y sin contraseña.
# Traemos solo 5 registros para ver cómo viene la información.
# 200 = conexión OK | 404 = no encontró | 500 = error del servidor
# ============================================================

# Dirección de la API
URL = "http://datos.energia.gob.ar/api/3/action/datastore_search"

# ID del dataset de Producción por Pozo (Dataset 1 - el principal)
RESOURCE_ID = "876b3746-85e2-4039-adeb-b1354436159f"

# Parámetros del pedido: qué dataset y cuántos registros
params = {
    "resource_id": RESOURCE_ID,
    "limit": 5
}

# Hacer el pedido a la API
print("⏳ Conectando a la API...")
respuesta = requests.get(URL, params=params)
print("📡 Código de respuesta:", respuesta.status_code)

if respuesta.status_code == 200:
    datos = respuesta.json()
    print("✅ Conexión exitosa!")
    print("📊 Total de registros en el dataset:", datos["result"]["total"])
    print("")
    print("--- Primeras 5 filas ---")
    
    # Convertir a DataFrame para verlo como tabla prolija
    df = pd.DataFrame(datos["result"]["records"])
    display(df)
else:
    print("❌ Error al conectar. Código:", respuesta.status_code)
```

```python
import requests

url = "http://datos.energia.gob.ar/api/3/action/package_list"
respuesta = requests.get(url)
datos = respuesta.json()

for dataset in datos["result"]:
    print(dataset)
```

```python
import requests

url = "http://datos.energia.gob.ar/api/3/action/package_list"
respuesta = requests.get(url)
lista = respuesta.json()["result"]

for nombre in lista:
    detalle = requests.get(f"http://datos.energia.gob.ar/api/3/action/package_show?id={nombre}").json()
    titulo = detalle["result"].get("title", "sin titulo")
    descripcion = detalle["result"].get("notes", "sin descripcion")
    print(f"DATASET: {titulo}")
    print(f"DESCRIPCION: {descripcion}")
    print("─" * 50)
```
